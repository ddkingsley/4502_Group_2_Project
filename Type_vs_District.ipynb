```python
%matplotlib inline
import numpy as np
import scipy as sp
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crimes = pd.read_csv('Train_Crime.csv', low_memory=False)
```

```python
bytype = crimes.groupby(['Primary Type', 'District']).size().reset_index(name='Count')
bytype.head()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(bytype['District'], bytype['Primary Type'], c=bytype['Count'], cmap=plt.cm.coolwarm)
```
