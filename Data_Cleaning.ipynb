```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Crimes_-_2001_to_present.csv', low_memory=False)
df.head()
```

```python
df.info()
```

```python
print(df.count()) # Counts the nonnull values for each column
```

```python
# The Ward and Community Area are being dropped due to the high percentage of null values.
# Other unneeded information includes the coordinates and when the information was updated.
df = df.drop(['Ward', 'Community Area','X Coordinate', 'Y Coordinate', 'Updated On', 'Latitude', 'Longitude', 'Location'], axis=1)
df.head()
```

```python
null_df = df[df.isnull().any(axis=1)]
print(null_df)
```

```python
null_df.count()
```

```python
null_df['Description'].value_counts()
```

```python
# We are not particularly interested in identity theft, so dropping the rest of the rows with null values will
# not be an issue for our project going forward.
df = df.dropna()
df = df.reset_index(drop=True)
df.District = df.District.astype(int) # Convert District from float to int.
df.head()
```
